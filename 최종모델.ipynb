- KNN(평균값, 중앙값, 최빈값 보다 정확하지만 이상치에 민감)

# 인코딩
- label encoding: 0부터 N까지의 숫자로 대체(머신러닝(특히 Decision Tree)에서는 숫자의 크기가 범위로 1과 0으로 나눌 때 영향을 줄 수 있음)
- ordinal encoding: 순서에 따라 0부터 N까지의 숫자로 대체 / 순서형 데이터에 적합(머신러닝(특히 Decision Tree)에서는 숫자의 크기가 범위로 1과 0으로 나눌 때 영향을 줄 수 있음)
- frequency encoding: 각 카테고리 값의 빈도수로 대체(머신러닝(특히 Decision Tree)에서는 숫자의 크기가 범위로 1과 0으로 나눌 때 영향을 줄 수 있음)
- one-hot encoding: 클래스 혹은 범주형(카테고리) 데이터의 갯수 만큼 컬럼이 늘어남 해당되는 카테고리 컬럼은 1, 그 외에는 0으로 대체
- target(mean, likelihood) encoding: one-hot encoding 처럼 각 카테고리 데이터에 의미있는 값으로 매핑하지만 컬럼은 늘어나지 않음 / 같은 카테고리 데이터의 target의 평균 값으로 대체 (과적합 가능성)
- leave one out encoding: 해당 값의 target 값을 제외하고 같은 카테고리에 있는 값의 target의 평균을 구해서 대체

## Name → Family
 - Name에서 성(Lastname) 추출

```python
def extract_lastname(data):
    
    families = []
    
    for i in range(len(data)):
        name = data.iloc[i]
        
        if '(' in name:
            name_no_bracket = name.split('(')[0]
        else: 
            name_no_bracket = name
            
        family = name_no_bracket.split(',')[0]
        title = name_no_bracket.split(',')[1].strip().split(' ')[0]
        
        for c in string.punctuation:
            family = family.replace(c, '').strip()
            
        families.append(family)
        
    return families

df_all['Family'] = extract_lastname(df_all['Name'])
df_train = df_all.loc[:890]
df_test = df_all.loc[891:]
dfs = [df_train, df_test]
```

- 여러번 폴드를 해서 0, 1 중 더 높은 확률의 레이블로 입력

```python
N = 5 # 폴드 수
oob = 0
probs = pd.DataFrame(np.zeros((len(X_test), N * 2)), columns=['Fold_{}_Prob_{}'.format(i, j) for i in range (1, N + 1) for j in range(2)])
```

```python
class_survived = [col for col in probs.columns if col.endswith('Prob_1')]

probs['1'] = probs[class_survived].sum(axis=1) / N
# 속성 이름이 'Prob_1'으로 끝나는 컬럼의 합을 폴드 개수로 나눔
probs['0'] = probs.drop(columns=class_survived).sum(axis=1) / N
# 속성 이름이 'Prob_1' 이외인 컬럼의 합을 폴드 개수로 나눔
probs['pred'] = 0
pos = probs[probs['1'] >= 0.5].index
probs.loc[pos, 'pred'] = 1

y_pred = probs['pred'].astype(int)
```
